#### Loop Current northen extension from February-May 2007.

#### Animation - Histograms with different number of bins
The animation confirms the statement above, that different number of bins leads to different and incoclusive results. 

```python
from numpy import histogram as nphistogram
from JSAnimation.IPython_display import display_animation, anim_to_html
from matplotlib import animation as animation


def getHistBinNumAni(data,totalframes=None,showpts=True):
    #ion()
    if totalframes is None:
        totalframes = min(len(data)-1,100)
    fig = figure()
    ax = fig.gca()

    n, bins, patches = plt.hist(data, totalframes, normed=1, facecolor='red', alpha=0.0)
    if showpts:
        junk = plt.plot(data,0.2*np.ones_like(data),'bo')
    def animate(i):
        n, bins = nphistogram(data, i+1, normed=False)
        #print n
        ax.set_ylim(0,1.1*n.max())
        ax.set_xlabel('Latitude')
        ax.set_title('Histograms with different bin sizes')
        for j in range(len(n)):
            rect,h = patches[j],n[j]
            #print h.max()
            x = bins[j]
            w = bins[j+1] - x
            rect.set_height(h)
            rect.set_x(x)
            rect.set_width(w)
            rect.set_alpha(0.75)
        #fig.canvas.draw()
    
    ani = animation.FuncAnimation(fig, animate, totalframes, repeat=False)
    return ani
```

```python
ani = getHistBinNumAni(x11)
display_animation(ani, default_mode='once')
```

## KDE (Kernel Density Estimation) 
is a helpful alternative to histograms. 


KDE, Gaussian kernel
--------------------

Using a Gaussian Kernel, we see that there are two picks around 26N and 28N which generally agrees with the results from the histograms. Although it is crude to say that the Loop Current northern extension lies mostly at these two latitudes, given that the sample is very small, the KDE is a good indicator about how far north the Loop Current mostly extends. 

```python
# Default of the KDEplot method in Seaborn
import seaborn as sns
sns.set(color_codes=True)
ax = sns.kdeplot(x11, shade=True, color="b")
ax.set_xlabel('Latitude')
```
